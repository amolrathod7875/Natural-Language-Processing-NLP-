````python
text = "India Launches new AI Policy"
tokens = word_tokenize(text)
stop_words = stopwords.words('english')
filtered_word = [word for word in tokens if word not in stop_words]
````

````python
analysis = TextBlob(text)
polarity, subjectivity = analysis.sentiment
if polarity > 0.05:
    print("Polarity : Positive")
elif polarity < -0.05:
    print("Polarity : Negative")
else:
    print("Polarity : Neutral")


if subjectivity > 0.6:
    print("subjectivity : Subjective/ Opinionated")
elif subjectivity < 0.4:
    print("subjectivity : Objective/ Factual")
else:
    print("subjectivity : Mixed")
````

````python
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
for text in doc :
    print(text, "-->", text.tag_,"-->",spacy.explain(text.tag_))
````

````python
texts = [['india','ai','policy'],['technology','growth']]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(t) for t in texts]
lda = LdaModel(corpus,num_topics=2,id2word=dictionary)
print(lda.print_topics())
````
