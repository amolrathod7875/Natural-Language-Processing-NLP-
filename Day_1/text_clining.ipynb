```python
def cleanhtml(sentence):
    #removes the word such as script and style from the sentence
    text = re.sub(r'(?is)<(script|style).*?>.*?</\1>','',sentence)
    #removes those html comments
    text = re.sub(r'(?s)<!--.*?-->','',text)
    #replacing rest all tags with space
    text = re.sub(r'<[^>]+>', " ", text) 
    #decode html entities 
    text = html.unescape(text)
    #replacing those extra space, line  with single space 
    text = re.sub(r'\s+',' ',text)
    return text
```

```python
cleanhtml_data = cleanhtml(sentence)
```

```python
def cleanwords(data):
    #lowering all the words into lower case 
    text = data.lower()
    #strip the punctuations
    text = re.sub(r'[^\w\s]','',text)
    #stopwords in english language
    stop_words = set(stopwords.words('english'))
    #tokens of the words 
    tokens = word_tokenize(text)
    #extracting the filterde words which are not in english stopwords
    filtered_words = [word for word in tokens if  word not in stop_words]
    return filtered_words
```

```python
filtered_data = cleanwords(cleanhtml_data)
```
